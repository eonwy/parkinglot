# 서울 (구)별 민영주차장 개소

```python
import json
import folium
import pandas as pd
import numpy as np
```

```python
geo_path = 'data/skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
geo_str
```

```python
dongData = pd.read_csv('data/guparkingcountprivate.csv', encoding='utf-8')
dongData
```

```python
dongData.dropna(subset=['개소'])
```

```python
dongData.fillna(method='pad', inplace=True)

dongData
```

```python
guparkingcountprivate = folium.Map(location=[37.5502, 126.982], zomm_start=10.5)
guparkingcountprivate
```

```python
folium.Choropleth(
    geo_data = geo_str, 
    data = dongData, 
    columns = ['자치구', '개소'], 
    fill_color = 'OrRd', 
    key_on = 'feature.properties.name'
    ).add_to(guparkingcountprivate)

guparkingcountprivate
```

```python
guparkingcountprivate.save('guparkingcountprivate.html')
guparkingcountprivate
```
