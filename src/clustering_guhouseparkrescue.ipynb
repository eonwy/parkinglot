# 서울 (구)별 주택가 주차장 확보율

```python
import json
import folium
import pandas as pd
import numpy as np
```

```python
geo_path = 'data/skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
geo_str
```

```python
dongData = pd.read_csv('data/guparkrescue.csv', encoding='utf-8')
dongData
```

```python
dongData.dropna(subset=['주택가주차장확보율'])
```

```python
dongData.fillna(method='pad', inplace=True)

dongData
```

```python
guhouparkrescuese = folium.Map(location=[37.5502, 126.982], zomm_start=10.5)
guhouparkrescuese
```

```python
folium.Choropleth(
    geo_data = geo_str, 
    data = dongData, 
    columns = ['행정자치구', '주택가주차장확보율'], 
    fill_color = 'OrRd', 
    key_on = 'feature.properties.name'
    ).add_to(guhouparkrescuese)

guhouparkrescuese
```

```python
guhouparkrescuese.save('guhouparkrescuese.html')
guhouparkrescuese
```
