# 서울 (구)별 가구수

```python
import json
import folium
import pandas as pd
import numpy as np
```

```python
geo_path = 'data/skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
geo_str
```

```python
dongData = pd.read_csv('data/guhousecount.csv', encoding='utf-8')
dongData.head()
dongData = dongData.groupby('자치구')['계'].sum()
dongData
```

```python
dongData.fillna(method='pad', inplace=True)

dongData
```

```python
guhousecount = folium.Map(location=[37.5502, 126.982], zomm_start=10.5)
guhousecount
```

```python
folium.Choropleth(
    geo_data = geo_str, 
    data = dongData, 
    columns = ['자치구', '계'], 
    fill_color = 'OrRd', 
    key_on = 'feature.properties.name'
    ).add_to(guhousecount)

guhousecount
```

```python
guhousecount.save('guhousecount.html')

guhousecount
```
