# 서울 (구)별 유입인구

```python
import json
import folium
import pandas as pd
import numpy as np
```

```python
geo_path = 'data/skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
geo_str
```

```python
dongData = pd.read_csv('data/gu_in.csv', encoding='utf-8')
dongData
```

```python
dongData.dropna(subset=['유입인구'])
```

```python
dongData.fillna(method='pad', inplace=True)

dongData
```

```python
gu_in = folium.Map(location=[37.5502, 126.982], zomm_start=10.5)
gu_in
```

```python
folium.Choropleth(
    geo_data = geo_str, 
    data = dongData, 
    columns = ['자치구', '유입인구'], 
    fill_color = 'OrRd', 
    key_on = 'feature.properties.name'
    ).add_to(gu_in)

gu_in
```

```python
gu_in.save('gu_in.html')
gu_in
```
